```python
answers = {} # создадим словарь для ответов


# Setting some Pandas params for easier work
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_rows', 50)


## Function that returns short name of the month and full name of the season by month number
# or full name of the season by short name of the month
# or list of short month names by season full name
def month_and_season(inp):
    months = [
        ["Jan", "winter"],
        ["Feb", "winter"],
        ["Mar", "spring"],
        ["Apr", "spring"],
        ["May", "spring"],
        ["Jun", "summer"],
        ["Jul", "summer"],
        ["Aug", "summer"],
        ["Sep", "autumn"],
        ["Oct", "autumn"],
        ["Nov", "autumn"],
        ["Dec", "winter"]
    ]
    if isinstance(inp, int):
        return months[inp-1]
    else:
        # Trying to find the month
        for item in months:
            if item[0]==inp:
                # returning name of the season
                return item[1]
        # if mnum is not a short month name, we probably got a season name ->
        # find all names of the months for the season
        season_months = []
        for item in months:
            if item[1]==inp:
                # adding month short name to the list to return
                season_months.append(item[0])
        return season_months


# Creating additional columns I'll need later
data["profit"] = data.revenue - data.budget
data["profitable"] = data.profit.apply(lambda x: x>0)
data["original_title_charnum"] = data.original_title.apply(lambda x: len(x))
data["overview_wordnum"] = data.overview.apply(lambda x: len(x.split(" ")))
data["genres_str"] = data.genres
data.genres = data.genres.apply(lambda x: x.split("|"))
data["director_str"] = data.director
data.director = data.director.apply(lambda x: x.split("|"))
data["cast_str"] = data.cast
data.cast = data.cast.apply(lambda x: x.split("|"))
data["production_companies_str"] = data.production_companies
data.production_companies = data.production_companies.apply(lambda x: x.split("|"))

# One way to extract month
data["release_month"] = data.release_date.apply(lambda x: month_and_season(int(x.split("/")[0]))[0])

data["release_date"] = pd.to_datetime(data["release_date"])

# Another way to extract month (working with dates, rather than strings):
from datetime import datetime as dt
data["release_month"] = data["release_date"].dt.month.apply(lambda x: month_and_season(x)[0])

data["release_season"] = data["release_month"].apply(lambda x: month_and_season(x)[1])
```

```python
# Solution 2

# Didn't put the funtion to the header as this is the only place I need it
def create_list_of_pairs(inlist):
    outlist = []
    for item in inlist:
        for item2 in inlist:
            if (item+" & "+item2 in outlist) or (item2+" & "+item in outlist) or item==item2:
                continue
            else:
                outlist.append(item+" & "+item2)
    return outlist

data["actors_pairs"] = data.cast.apply(create_list_of_pairs)
data.explode("actors_pairs").groupby("actors_pairs").count().sort_values(by="imdb_id", ascending=False).imdb_id
```
